```python
# 1,2
class vehicle:
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name_of_vehicle=name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
    
```

```python
class car(vehicle):
    def seating_capacity(self,capacity):
        self.capacity = capacity
        return self.name_of_vehicle,self.capacity
    
```

```python
vh = car('maruti','100kph','2000km')
```

```python
vh.seating_capacity(5)
```

```python
#3.Multiple Inheritance
class a:
    def func1(self):
        print("func1 method of class a")
class b:
     def func2(self):
        print("func2 method of class b")
class c(a,b):
     pass
c_obj=c()
c_obj.func1()
c_obj.func2()
```

```python
#4
class pwskills:
    def set_count(self,count):
        self.__cnt = count
    def get_count(self):
        return self.__cnt
p = pwskills()
```

```python
p.set_count(5000)
```

```python
p.get_count()
```

```python
#5 Method Overriding
class animal:
    def eat(self):
        print("Animal is eating")
    def sleep(self):
        print("Animal is sleeping")
class dog(animal):
     def eat(self):
        print("Dog is eating")
d = dog()
d.eat()
```
